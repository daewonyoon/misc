```python
# Summarization Using TextRank (by lovit)
```

```python
from textrank import KeysentenceSummarizer
from konlpy.tag import Okt

okt = Okt()

summarizer = KeysentenceSummarizer(
    tokenize = okt.pos,
    min_sim = 0.5,
    verbose = True
)

keysents = summarizer.summarize(train.iloc[0, 2], topk=3)
```

```python
from typing import List
from tqdm.notebook import tqdm

okt = Okt()
summarizer = KeysentenceSummarizer(
    tokenize = okt.pos,
    min_sim = 0.5,
    verbose = False
)

def extract_keysent(sents:List[str])->str:      
    keysents = summarizer.summarize(sents, topk=3)
    keysents.sort()
    return '\n'.join(ks[2] for ks in keysents)

tqdm.pandas()
test["summary"] = test.article_original.progress_apply(extract_keysent)    
```

```python
import numpy as np
import pandas as pd
import scipy as sp
import scipy.stats
import seaborn as sns

import matplotlib.pyplot as plt

plt.rcParams['font.family'] = 'D2Coding'
```

```python
mu, std = 0, 1
rv = sp.stats.norm(mu, std)
```

```python
xx = np.linspace(-5, 5, 100)
plt.grid(True)
plt.plot(xx, rv.pdf(xx))

plt.arrow(0, 0.05, 2, 0, lw=3, color='r', head_width=0.02, head_length=0.2, length_includes_head=True)
plt.arrow(0, 0.05, -2, 0, lw=3, color='r', head_width=0.02, head_length=0.2, length_includes_head=True)
plt.text(-0.95, -.03, "표준편차의 약 4배")
```

```python
np.random.seed(0)
x = rv.rvs(20)
x
```

```python
ns, _ = np.histogram(x)
plt.grid(True)
plt.bar(x=list(range(10)), height=ns)
```

```python
sns.distplot(x, rug=True, fit=sp.stats.norm)
```

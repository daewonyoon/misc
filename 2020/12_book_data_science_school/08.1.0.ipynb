```python
import numpy as np
import matplotlib.pyplot as plt
import scipy as sp

```

```python
xx = np.linspace(-1, 1, 199)
rv = sp.stats.norm(loc=0, scale=0.1)
pdf = rv.pdf(xx)
plt.grid(True)
plt.plot(xx, pdf)
plt.title("gaussian distribution")
plt.xlabel("$x$")
plt.ylabel("$p(x)$")
plt.show()
```

```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
```

```python

iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df
```

```python
sy = pd.Series(iris.target, dtype="category")
sy = sy.cat.rename_categories(iris.target_names)
df["species"] = sy
```

```python
df
```

```python
# virginica 꽃잎 길이에 대해 표본평균, 표본중앙값, 표본최빈값을 구하고 그림을 그려라.
# 표본최빈값을 구할 때에는 1cm 간격으로 히스토그램을 그려라.
# petal 꽃잎, sepal 꽃받침

virg_petal_len_df = df[df["species"] == "virginica"].iloc[:,2:3]

virg_petal_len_df.describe()
```

```python

virg_petal_len = df[df["species"] == "virginica"].iloc[:,2]
```

```python
print(f"mean = {virg_petal_len.mean()}, median = {virg_petal_len.median()}")
```

```python
print(virg_petal_len.min(), virg_petal_len.max())
```

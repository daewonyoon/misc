```python
import pandas as pd
from sklearn.datasets import load_iris

iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df
```

```python
sy = pd.Series(iris.target, dtype="category")
sy = sy.cat.rename_categories(iris.target_names)
df["species"] = sy
```

```python
df
```

```python
# 꽃잎의 길이는 단봉분포인가 다봉분포인가?
# petal 꽃잎, sepal 꽃받침

df.iloc[:,2:3].plot.hist(bins=20)
```

```python
# setosa 인 데이터만 뽑아서 히스토그램을 그려라. 
# 단봉분포인가 다봉분포인가?

df[df["species"]=="setosa"].iloc[:,2:3].plot.hist()
```

```python
# 나의 연습, 스캐터 플롯

df.plot.scatter(x=0, y=2, c=iris.target)
```

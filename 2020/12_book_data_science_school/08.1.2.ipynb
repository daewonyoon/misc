```python
import math

import numpy as np
import matplotlib.pyplot as plt
import scipy as sp
import seaborn as sns

```

```python
rv = sp.stats.norm()
```

```python
samples = rv.rvs(size=1000, random_state=0)

plt.grid(True)
sns.distplot(samples)
plt.title("normal distribution sample")
```

```python
samples.mean()
```

```python
std_unb, std_b = samples.std(ddof=1), samples.std(ddof=0)
l = len(samples)
print(f'unbiased std ={std_unb}, biased std ={std_b}, biased x root(N/(N/1) ={std_b*math.sqrt(l/(l-1))}')
```
